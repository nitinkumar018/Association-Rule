# Association Rule Mining:

# Implement an Apriori algorithm 

```python
# Market Basket Analysis of Online retail data

# minimum support = 45%, threshold confidence = 80% 
# lift says how likely item Y is purchased when item X is purchased; lift > 1 means item Y is likely to bought if X is bought 
# min_length means minimum no. of items in a rule 
from apyori import apriori #library apyori is used to calculate the association rule using function Apriori
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
```

```python
association_rules
```

```python
association_results = list(association_rules) # it will calculate all the association rules/ relations between the items 
```

```python
association_results
```

```python
association_results[0]
```

```python
len(association_results)  # no. of association rules/ relations in the dataset
```

```python
print("There are {} Relation derived.".format(len(association_results))) # there are 34 association rules/ relations b/w the items
 
```

```python
#34 association rules derived/ generated
for i in range(0, len(association_results)):
    print(association_results[i][0])
```

```python
# calculate support value, threshold confidence value for each association rule / relation
for item in association_results:
    # first index of the inner list
    # Contains base item and add item
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + str(items[1:]))

    # second index of the inner list
    print("Support: " + str(item[1]))

    # third index of the list located at 0th
    # of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```

# Interview Questions:

1. What is lift and why is it important in Association rules?

The lift value is an important measure in association rule mining because it provides a way to distinguish between significant and insignificant association rules. A high lift value indicates that the association rule is more significant, as it suggests that the two items are highly dependent on each other

2.What is support and Confidence. How do you calculate them?

Support is a measure that gives an idea of how frequent an itemset is in all the transactions. Confidence measures the likelihood of items given that the shopping cart already has other items

Support = no. of transactions a particular item occurs / total no. of transactions 

Ex: apple occurs in 3 transactions out of total 4 transactions, then support = 3/4 = 0.75 i.e. 75% 


3 What are some limitations or challenges of Association rules mining?

challenge in association rule mining is to set the appropriate thresholds for the metrics that are used to generate and evaluate the rules. The thresholds determine how many and which rules are extracted from the data.
